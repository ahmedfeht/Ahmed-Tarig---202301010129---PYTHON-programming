WEEK 3 - TUTORIAL 3

Inputs:
- Quiz 1 mark
- Quiz 2 mark
- Quiz 3 mark

Process:
1. Enter three quiz marks.
2. Calculate the average mark.
3. Display the average mark.
4. Determine PASS or FAIL.
5. Ask user whether to continue.

Outputs:
- Average Mark
- PASS or FAIL
- Program Ended

Pseudocode:

START

Set choice = "y"

WHILE choice = "y"
    Input quiz_1
    Input quiz_2
    Input quiz_3

    average = (quiz_1 + quiz_2 + quiz_3) / 3

    Display average

    IF average >= 50
        Display PASS
    ELSE
        Display FAIL
    ENDIF

    Input choice

ENDWHILE

Display Program Ended

END

Algorithm Structure:
- Sequence
- Selection
- Iteration
